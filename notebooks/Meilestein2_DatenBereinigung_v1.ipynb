```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
pd.set_option("display.max_columns", None)

CSV_PATH = "../en.openfoodfacts.org.products 2.csv"

df = pd.read_csv(CSV_PATH, sep="\t", low_memory=False, on_bad_lines="skip")

#Ozan: Bereinigung der Spalten brands, labels, countries, food_groups und main_category

#Brands
brands_completeness = str(df["brands"].notna().mean()*100)
brands_before = df["brands"].notna().sum()

print("Vor Zusammenführen war Spalte zu " + brands_completeness + " gefüllt")

#brands behalten, mit brands_en auffüllen falls leer, mit brands_tags auffüllen falls auch leer
df["brands"] = df["brands"].fillna(df["brands_en"].fillna(df["brands_tags"]))

#Prozent und Anzahl der Spalte nach Bearbeitung
brands_completeness = str(df["brands"].notna().mean()*100)
labels_en_after = df["brands"].notna().sum()

countries_added = str(labels_en_after - brands_before)

print("Nach Zusammenführen ist Spalte zu " + brands_completeness + " gefüllt")
print("Es wurden " + countries_added + " Einträge hinzugefügt")

#brands löschen
df = df.drop(columns=["brands_tags", "brands_en"])


#Labels
labels_en_completeness = str(df["labels_en"].notna().mean()*100)
labels_en_before = df["labels_en"].notna().sum()

print("Vor Zusammenführen war Spalte zu " + labels_en_completeness + " gefüllt")

#labels_en mit labels auffüllen falls leer, mit labels_tags auffüllen falls auch leer
df["labels_en"] = (df["labels_en"]
                                .fillna(df["labels"]
                                .fillna(df["labels_tags"])))

#Prozent und Anzahl der Spalte nach Bearbeitung
labels_en_completeness = str(df["labels_en"].notna().mean()*100)
labels_en_after = df["labels_en"].notna().sum()

labels_added = str(labels_en_after - labels_en_before)

print("Nach Zusammenführen ist Spalte zu " + labels_en_completeness + " gefüllt")
print("Es wurden " + labels_added + " Einträge hinzugefügt")

#labels und labels_tags löschen
df = df.drop(columns=["labels", "labels_tags"])


#Countries
countries_en_completeness = str(df["countries_en"].notna().mean()*100)
countries_en_before = df["countries_en"].notna().sum()

print("Vor Zusammenführen war Spalte zu " + countries_en_completeness + " gefüllt")

#countries_en mit countries_tags auffüllen falls leer, mit countries auffüllen falls auch leer
df["countries_en"] = (df["countries_en"]
                                .fillna(df["countries_tags"]
                                .fillna(df["countries"])))

#Prozent und Anzahl der Spalte nach Bearbeitung
countries_en_completeness = str(df["countries_en"].notna().mean()*100)
countries_en_after = df["countries_en"].notna().sum()

countries_added = str(countries_en_after - countries_en_before)

print("Nach Zusammenführen ist Spalte zu " + countries_en_completeness + " gefüllt")
print("Es wurden " + countries_added + " Einträge hinzugefügt")

#countries und countries_tags löschen
df = df.drop(columns=["countries", "countries_tags"])

#Food Groups
food_groups_en_completeness = str(df["food_groups_en"].notna().mean()*100)
food_groups_en_before = df["food_groups_en"].notna().sum()

print("Vor Zusammenführen war Spalte zu " + food_groups_en_completeness + " gefüllt")

#food_groups_en mit food_groups_tags auffüllen falls leer,
#mit food_groups auffüllen falls auch leer
df["food_groups_en"] = (df["food_groups_en"]
                                .fillna(df["food_groups_tags"]
                                .fillna(df["food_groups"])))

#Prozent und Anzahl der Spalte nach Bearbeitung
food_groups_en_completeness = str(df["food_groups_en"].notna().mean()*100)
food_groups_en_after = df["food_groups_en"].notna().sum()

food_groups_en_added = str(food_groups_en_after - food_groups_en_before)

print("Nach Zusammenführen ist Spalte zu " + food_groups_en_completeness + " gefüllt")
print("Es wurden " + food_groups_en_added + " Einträge hinzugefügt")

#food_groups und food_groups_tags löschen
df = df.drop(columns=["food_groups", "food_groups_tags"])

#Main Categories
main_category_en_completeness = str(df["main_category_en"].notna().mean()*100)
main_category_en_before = df["main_category_en"].notna().sum()

print("Vor Zusammenführen war Spalte zu " + main_category_en_completeness + " gefüllt")

#main_categories_en mit main_category auffüllen falls leer
df["main_category_en"] = (df["main_category_en"]
                                .fillna(df["main_category"]))

#Prozent und Anzahl der Spalte nach Bearbeitung
main_category_en_completeness = str(df["main_category_en"].notna().mean()*100)
main_category_en_after = df["main_category_en"].notna().sum()

main_category_added = str(main_category_en_after - main_category_en_before)

print("Nach Zusammenführen ist Spalte zu " + main_category_en_completeness + " gefüllt")
print("Es wurden " + main_category_added + " Einträge hinzugefügt")

#main_category löschen
df = df.drop(columns=["main_category"])

#Meme

# Defragmentierung des DataFrames bevor neue Spalten hinzugefügt werden
df = df.copy()

#Packaging
df['packaging_en'] = (
    df['packaging_en']
    .fillna(df['packaging'])
    .fillna(df['packaging_tags'])
    .fillna(df['packaging_text'])
)
df = df.drop(columns=['packaging', 'packaging_tags', 'packaging_text'])

#Additives
df['additives_tags'] = (
    df['additives_tags']
    .fillna(df['additives_en'])
)
df = df.drop(columns=['additives', 'additives_en'])

#Ingredients
df['ingredients_tags'] = (
    df['ingredients_tags']
    .fillna(df['ingredients_text'])
)
df = df.drop(columns=['ingredients_text'])

#Spalten umbenennen
df = (df
                   .rename(columns={"labels_en" : "labels"
                                    , "countries_en" : "countries"
                                    , "food_groups_en" : "food_groups"
                                    , "main_category_en" : "main_category"
                                    , "packaging_en" : "packaging"
                                    , "additives_tags" : "additives"
                                    , "ingredients_tags" : "ingredients"
                                    }))
```

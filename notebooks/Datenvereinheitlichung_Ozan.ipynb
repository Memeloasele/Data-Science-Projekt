## Dieses Notebook vereinheitlicht Inhalte einer Spalte
Dafür werden:
-  Einträge in lowercase umgewandelt
-  Präfixe entfernt
-  Mehrere Einträge in einer Zelle als Liste der separaten Einträge erstellt
# Spalten
-  brands
-  labels
-  countries
-  food_groups
-  main_category
-  pnns_groups_1
-  pnns_groups_2
-  brand_owner
-  popularity_tags

## Mögliche Probleme
# Spalten
-  labels hat Nutriscore drinne, aber einmal A und einmal B --> mit Nutriscore abgleichen und beheben

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 100)
```

```python
df_original = pd.read_csv("openfoodfacts_ueber6prozent.csv.zip", sep="\t", low_memory=False, on_bad_lines="skip")
```

```python
df = df_original.copy()
```

```python
spalten = ["brands", "labels", "countries", "food_groups", 
            "main_category", "pnns_groups_1", "pnns_groups_2", 
            "brand_owner", "popularity_tags"]

# Wandle Einträge in Kleinbuchstaben um und entferne Leerzeichen vorne und hinten
df[spalten] = df[spalten].apply(lambda x: x.str.lower().str.strip())
```

# Anmerkung
Alles wird in Listen umgewandelt, auch einzelne Einträge. Das ist best-practise da man sonst beim Weiterarbeiten den Code so anpassen müsste, dass er zuerst überprüfen muss ob ein Eintrag ein String oder eine Liste beinhaltet, was alles komplizierter machen würde

```python
# Einträge in Listen umwandeln
# ACHTUNG: Wenn man den Teil zweimal ausführt werden alle Einträge zu NaN da Listen keine strings sind
df[spalten] = df[spalten].apply(lambda x: x.str.split(r",\s*"))
```

Alle Spalteneinträge in lowercase umgewandelt und in Listen überführt

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
pd.set_option("display.max_columns", None)
```

```python
data = pd.read_csv("Bereinigung.csv", sep="\t", low_memory=False, on_bad_lines="skip")
```

```python
data_bereinigen = data.copy()
```

### Brands Ergebnis
-  Vollständigkeit von 62,73% auf 62,74% Prozent
-  839 Einträge hinzugefügt in brands Spalte
-  brands_tags und brands_en gelöscht

```python
#Prozent und Anzahl der Spalte vor Bearbeitung
brands_completeness = str(data_bereinigen["brands"].notna().mean()*100)
brands_before = data_bereinigen["brands"].notna().sum()

print("Vor Zusammenführen war Spalte zu " + brands_completeness + " gefüllt")

#brands behalten, mit brands_en auffüllen falls leer, mit brands_tags auffüllen falls auch leer
data_bereinigen["brands"] = data_bereinigen["brands"].fillna(data_bereinigen["brands_en"].fillna(data_bereinigen["brands_tags"]))

#Prozent und Anzahl der Spalte nach Bearbeitung
brands_completeness = str(data_bereinigen["brands"].notna().mean()*100)
labels_en_after = data_bereinigen["brands"].notna().sum()

countries_added = str(labels_en_after - brands_before)

print("Nach Zusammenführen ist Spalte zu " + brands_completeness + " gefüllt")
print("Es wurden " + countries_added + " Einträge hinzugefügt")

#brands löschen
data_bereinigen = data_bereinigen.drop(columns=["brands_tags", "brands_en"])
```

### Labels Ergebnis
-  Vollständigkeit von 27,2351% auf 27,2359% Prozent gestiegen
-  38 Einträge labels_en Spalte hinzugefügt
-  labels und labels_tags gelöscht

```python
#Prozent und Anzahl der Spalte vor Bearbeitung
labels_en_completeness = str(data_bereinigen["labels_en"].notna().mean()*100)
labels_en_before = data_bereinigen["labels_en"].notna().sum()

print("Vor Zusammenführen war Spalte zu " + labels_en_completeness + " gefüllt")

#labels_en mit labels auffüllen falls leer, mit labels_tags auffüllen falls auch leer
data_bereinigen["labels_en"] = (data_bereinigen["labels_en"]
                                .fillna(data_bereinigen["labels"]
                                .fillna(data_bereinigen["labels_tags"])))

#Prozent und Anzahl der Spalte nach Bearbeitung
labels_en_completeness = str(data_bereinigen["labels_en"].notna().mean()*100)
labels_en_after = data_bereinigen["labels_en"].notna().sum()

labels_added = str(labels_en_after - labels_en_before)

print("Nach Zusammenführen ist Spalte zu " + labels_en_completeness + " gefüllt")
print("Es wurden " + labels_added + " Einträge hinzugefügt")

#labels und labels_tags löschen
data_bereinigen = data_bereinigen.drop(columns=["labels", "labels_tags"])
```

### Countries Ergebnisse
-  Vollständigkeit eigentlich gleich geblieben mit 99,49%
-  5 Einträge in countries_en Spalte aufgefüllt
-  countries und countries_tags Spalte gelöscht

```python
#Prozent und Anzahl der Spalte vor Bearbeitung
countries_en_completeness = str(data_bereinigen["countries_en"].notna().mean()*100)
countries_en_before = data_bereinigen["countries_en"].notna().sum()

print("Vor Zusammenführen war Spalte zu " + countries_en_completeness + " gefüllt")

#countries_en mit countries_tags auffüllen falls leer, mit countries auffüllen falls auch leer
data_bereinigen["countries_en"] = (data_bereinigen["countries_en"]
                                .fillna(data_bereinigen["countries_tags"]
                                .fillna(data_bereinigen["countries"])))

#Prozent und Anzahl der Spalte nach Bearbeitung
countries_en_completeness = str(data_bereinigen["countries_en"].notna().mean()*100)
countries_en_after = data_bereinigen["countries_en"].notna().sum()

countries_added = str(countries_en_after - countries_en_before)

print("Nach Zusammenführen ist Spalte zu " + countries_en_completeness + " gefüllt")
print("Es wurden " + countries_added + " Einträge hinzugefügt")

#countries und countries_tags löschen
data_bereinigen = data_bereinigen.drop(columns=["countries", "countries_tags"])
```

### Food_groups Ergebnisse
-  Vollständigkeit eigentlich auf 32,21% geblieben
-  1 Eintrag wurde food_groups_en aufgefüllt
-  food_groups und food_groups_tags Spalte gelöscht

```python
#Prozent und Anzahl der Spalte vor Bearbeitung
food_groups_en_completeness = str(data_bereinigen["food_groups_en"].notna().mean()*100)
food_groups_en_before = data_bereinigen["food_groups_en"].notna().sum()

print("Vor Zusammenführen war Spalte zu " + food_groups_en_completeness + " gefüllt")

#food_groups_en mit food_groups_tags auffüllen falls leer, 
#mit food_groups auffüllen falls auch leer
data_bereinigen["food_groups_en"] = (data_bereinigen["food_groups_en"]
                                .fillna(data_bereinigen["food_groups_tags"]
                                .fillna(data_bereinigen["food_groups"])))

#Prozent und Anzahl der Spalte nach Bearbeitung
food_groups_en_completeness = str(data_bereinigen["food_groups_en"].notna().mean()*100)
food_groups_en_after = data_bereinigen["food_groups_en"].notna().sum()

food_groups_en_added = str(food_groups_en_after - food_groups_en_before)

print("Nach Zusammenführen ist Spalte zu " + food_groups_en_completeness + " gefüllt")
print("Es wurden " + food_groups_en_added + " Einträge hinzugefügt")

#food_groups und food_groups_tags löschen
data_bereinigen = data_bereinigen.drop(columns=["food_groups", "food_groups_tags"])
```

### Main_category Ergebnisse
-  Vollständigkeit eigentlich mit 41,09% gleich geblieben
-  2 Einträge in main_category_en Spalte aufgefüllt
-  main_category Spalte gelöscht

```python
#Prozent und Anzahl der Spalte vor Bearbeitung
main_category_en_completeness = str(data_bereinigen["main_category_en"].notna().mean()*100)
main_category_en_before = data_bereinigen["main_category_en"].notna().sum()

print("Vor Zusammenführen war Spalte zu " + main_category_en_completeness + " gefüllt")

#main_categories_en mit main_category auffüllen falls leer
data_bereinigen["main_category_en"] = (data_bereinigen["main_category_en"]
                                .fillna(data_bereinigen["main_category"]))

#Prozent und Anzahl der Spalte nach Bearbeitung
main_category_en_completeness = str(data_bereinigen["main_category_en"].notna().mean()*100)
main_category_en_after = data_bereinigen["main_category_en"].notna().sum()

main_category_added = str(main_category_en_after - main_category_en_before)

print("Nach Zusammenführen ist Spalte zu " + main_category_en_completeness + " gefüllt")
print("Es wurden " + main_category_added + " Einträge hinzugefügt")

#main_category löschen
data_bereinigen = data_bereinigen.drop(columns=["main_category"])
```

```python
#Spalten umbenennen
data_bereinigen = (data_bereinigen
                   .rename(columns={"labels_en" : "labels"
                                    , "countries_en" : "countries"
                                    , "food_groups_en" : "food_groups"
                                    , "main_category_en" : "main_category"}))
```

### Entferne Präfixes aus _tags Spalten


```python
spalten_zu_bearbeiten = ["labels", "countries", "food_groups", "brands", "main_category"]

#entferne en: Präfix
data_bereinigen[spalten_zu_bearbeiten] = (data_bereinigen[spalten_zu_bearbeiten]
                                          .apply(lambda x: x.str.strip()
                                                 .str.removeprefix("en:")))

#entferne xx: Präfix
data_bereinigen[spalten_zu_bearbeiten] = (data_bereinigen[spalten_zu_bearbeiten]
                                          .apply(lambda x: x.str.strip()
                                                 .str.removeprefix("xx:")))

#ersetze "-" mit Leerzeichen um z.B. unites-states zu unites states zu machen
data_bereinigen[spalten_zu_bearbeiten] = (data_bereinigen[spalten_zu_bearbeiten]
                                          .apply(lambda x: x.str.strip()
                                                 .str.replace("-", " ", regex=False)))
```
